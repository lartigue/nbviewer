```python
import numpy as np
import scipy.interpolate as interpolate
from scipy import fftpack
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# frquence dchantillonnage en Hz
fe = 100
# dure en seconde
T = 10
# Nombre de point :
N = T*fe
# Array temporel :
t = np.arange(0.,N+1)/fe
# frquence du signal : Hz
f0 = 2
# signal temporel
sinus = np.sin(2*np.pi*f0*t)
# signal frquentiel : on divise par la taille du vecteur pour normaliser la fft
fourier = fftpack.fft(sinus)/np.size(sinus)
# axe frquentiel:
axe_f = np.arange(0.,N+1)*fe/N
# interpolation
F_sinus = interpolate.interp1d(t,sinus)
ratio=1.1
t2=np.arange(0.,ratio*N+1)/(fe*ratio)
axe_f2 = np.arange(0.,ratio*N+1)*fe/N
sinus2 = F_sinus(t2)
fourier2 = fftpack.fft(sinus2)/np.size(sinus2)
# plot
plt.figure(figsize=(10, 4), dpi=80,)
plt1=plt.subplot(121)
plt.plot(t,sinus,'r-')
plt.xlabel('axe temporel, en seconde')
plt2=plt.subplot(122)
plt.plot(axe_f,np.abs(fourier),'x-')
#plt2.set_xscale('log')
plt2.set_yscale('log')
plt2.set_xlim([fe/(2*N),fe/2])
plt2.set_ylim([1e-6,1])
plt.xlabel('axe frequentiels en Hertz')
plt.show()
# plot
plt.figure(figsize=(10, 4), dpi=80,)
plt1=plt.subplot(121)
plt.plot(t2,sinus2,'r-')
plt.xlabel('axe temporel, en seconde')
plt2=plt.subplot(122)
plt.plot(axe_f2,np.abs(fourier2),'x-')
#plt2.set_xscale('log')
plt2.set_yscale('log')
plt2.set_xlim([fe/(2*N),fe/2])
plt2.set_ylim([1e-6,1])
plt.xlabel('axe frequentiels en Hertz')
```
